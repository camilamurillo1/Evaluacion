##Metodo de desviacion estandar para eliminar outliers del DF

```python
df5.info()
```

```python
filtroNum=df5.iloc[ : , [0, 2, 9, 22, 23, 30,31,34,37,38,41-48,51-54,56,58,70-74]]  #Columnas no consecutivas 
filtroNum
```

```python
filtroLetra=df5.iloc[:,[1,3-8,10-21,24-29,32,33,35,36,39,40,49,50,55,57-69]]
filtroLetra
```

```python
fig = plt.figure(figsize =(15, 8))
filtroNum.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=filtroNum
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
outliers= filtroNum[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df6= filtroNum[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df6
```

```python
valores_nulos=df6.isnull().sum()
valores_nulos
```

```python
filtroLetra
```

```python
df_clean=df6.copy()
df_clean=df_clean.fillna(00)
df_clean
```
