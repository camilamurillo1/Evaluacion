##Metodo de desviacion estandar para eliminar outliers del DF

```python
filtroNum1.info()
```

```python
fig = plt.figure(figsize =(15, 8))
filtroNum1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=filtroNum1
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
outliers= filtroNum1[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
df6= filtroNum1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df6
```

```python
valores_nulos=df6.isnull().sum()
valores_nulos
```

```python
df_clean=df6.copy()
df_clean=df_clean.fillna(round(df6.mean(),1))
df_clean
```

```python
valores_nulos=df_clean.isnull().sum().sum()
valores_nulos
```

```python
df= pd.concat([df_clean, filtroLetra], axis=1)
df.info()
```

```python
df.to_csv("NulosUSA.csv")

from google.colab import files 
   
files.download("NulosUSA.csv")
```
